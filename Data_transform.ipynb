```python
import pandas as pd
```

```python
amazon = pd.read_csv('Datasets_limpios/amazon_limpio.csv')
disney = pd.read_csv('Datasets_limpios/disney_limpio.csv')
hulu = pd.read_csv('Datasets_limpios/hulu_limpio.csv')
netflix = pd.read_csv('Datasets_limpios/netflix_limpio.csv')
```

```python
amazon['rating'] = amazon['rating'].str.lower()
disney['rating'] = disney['rating'].str.lower()
hulu['rating'] = hulu['rating'].str.lower()
netflix['rating'] = netflix['rating'].str.lower()
```

```python
amazon['platform'] = 'amazon'
disney['platform'] = 'disney'
hulu['platform'] = 'hulu'
netflix['platform'] = 'netflix'
```

```python
df = pd.concat([amazon,disney,hulu,netflix],axis=0)
df.sample(10)
```

```python
amazon.to_csv('amazon_limpio.csv',index=False)
disney.to_csv('disney_limpio.csv',index=False)
hulu.to_csv('hulu_limpio.csv',index=False)
netflix.to_csv('netflix_limpio.csv',index=False)
```

```python
df.to_csv('datos_completos.csv',index=False)
```
